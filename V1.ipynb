```python
file = open ("C:/Users/Flori/Pictures/IA/VisDrone2019-DET-test-dev/annotations/0000006_00159_d_0000001.txt", "r")
# file2 = open ("C:/Users/Flori/Pictures/IA/VisDrone2019-DET-test-dev/annotations/0000006_00159_d_0000002.txt", "r")
```

```python
lines = file.readlines()
# lines = file2.readlines()
# print (lines) 
file.close()

for line in lines:
    
    
#     print(line.strip())
    x = line.split(",", 8)
    #     print(x)
    
    x1 =  int(x[0])
    y1 =  int(x[1])
    x2 =  x1 +  int(x[2])
    y2 = y1 +  int(x[3])
    
    category = int(x[5])
    if (category == 1):
        color = (255, 255, 255)
    elif (category == 4):
        color = (0, 0, 255)
    elif (category == 11):
        color = (0, 255, 255)

    top_left_corner = (x1, y1)
    botom_left_corner =(x2, y2)
    
#     print(top_left_corner)
    
    thickness = 5
    photo_test = cv2.rectangle(photo_test, top_left_corner, botom_left_corner, color, thickness)
    
    
plt.imshow(photo_test)
```
